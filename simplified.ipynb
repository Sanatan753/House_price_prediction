```python
import pandas as pd
```

```python
data = pd.read_json('data.json') #Reading the data from json
data.head()
content = data['content'] #Extracting the content column as it's in a format of column of dictionaries 
# content is a Pandas Series now
data.head()
```

```python
content.values # it's a numpy array of dictionaries containing all values(dictionaries) from content

content_list = content.values.tolist() #Converting the np array to list

content_data = pd.DataFrame(content_list) #Converting the list to data frame

content_data['srp'] = data['srp'] #Adding the srp column from the original data frame to new data frame

content_data.set_index('srp',inplace=True) #setting srp as index 
#because it does have unique values for each row you can reset it if you don't want it as an index

final = content_data # Just for simplification and naming 
final #tadaaa
```
